```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv('~/Documents/Data Science Masterclass/DATA/Advertising.csv')
```

```python
df['expenditure'] = df['TV'] + df['radio'] + df['newspaper']
df
```

```python
sbn.scatterplot(data=df, x='expenditure', y='sales')
```

## Plotting Best Fit Line using Numpy

```python
sbn.regplot(data=df, x='expenditure', y='sales')
```

```python
X = df['expenditure']
```

```python
y = df['sales']
```

## Calculating Coefficients *a* & *b* using Ordinary Least Squared Method
* Solve for the equation *y = bX + a*
* Use np.polyfit()

```python
np.polyfit(X, y, deg=1)
```

i.e *y = 0.0486X + 4.2430*

```python
potential_expenditure = np.arange(0, 500)
```

```python
expected_sales = 0.04868788 * potential_expenditure + 4.24302822
```

```python
plt.figure(figsize=(12,6), dpi=100)
sbn.scatterplot(data=df ,x='expenditure', y='sales')
plt.plot(potential_expenditure, expected_sales, color='red')
plt.xlim(0)
plt.ylim(0)
```
