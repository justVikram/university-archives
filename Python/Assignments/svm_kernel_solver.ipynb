```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
operand_1 = pd.Series(np.array([1, -1, -1, 1]))
operand_2 = pd.Series(np.array([1, -1, 1, -1]))
result    = pd.Series(np.array([1, 1, -1, -1]))

dict = {
    'operand_1': operand_1,
    'operand_2': operand_2,
    'result': result,
}
```

```python
df = pd.DataFrame(dict)
df
```

```python
# Utility function to plot decision boundary

def plot_svm_boundary(model,X,y):
    
    X = X.values
    y = y.values
    
    plt.scatter(X[:, 0], X[:, 1], c=y, s=30,cmap='seismic')

    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])

    
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
               linewidth=1, facecolors='none', edgecolors='k')
    plt.show()
```

```python
# Visualizing data

plt.figure(dpi=100)
sbn.scatterplot(data=df, x='operand_1', y='operand_2', hue='result')
```

```python
y = df.get('result')
X = df.drop('result', axis=1)
```

```python
from sklearn.svm import SVC
```

# Using Linear Kernel

```python
model = SVC(kernel='linear', C=1)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=10)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='linear', C=1000)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

# Using RBF Kernel

Model starts overfitting for C = 10 and C = 1000

```python
model = SVC(kernel='rbf', C=1, gamma='scale')
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='rbf', C=10, gamma='scale')
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='rbf', C=1000, gamma='scale')
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

# Using Sigmoid Kernel

```python
model = SVC(kernel='sigmoid', C=1)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='sigmoid', C=10)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='sigmoid', C=1000)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```
