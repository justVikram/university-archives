```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image, ImageFilter
```

```python
image = Image.open('fg.jpg')
image = image.convert("L")
image = image.filter(ImageFilter.FIND_EDGES)
plt.imshow(image)
image.save('spiderman.png')
```

```python
background_image = Image.open('bg.jpg')
foreground_image = Image.open('spiderman.png')

foreground_image = foreground_image.resize((300, 600))
```

```python
temp_bg = background_image.copy()
temp_bg.paste(foreground_image, (400, 0))
plt.imshow(temp_bg)
```
