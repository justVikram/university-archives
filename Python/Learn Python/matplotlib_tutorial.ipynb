## Functional Programming Method

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
plt.plot(x, y)
plt.title('Y vs X')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(left=0)
plt.ylim(bottom=0)
#plt.savefig('myfirstplot.png', bbox_inches='tight') # Save the generated PNG
```

## Object-oriented Programming Method
(For more control over the plots)

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x,y)
# first two values specify origin; next two specify width and height of axes wrt the figure (1:1 is optimum)
```

### Making a sample plot

```python
a = np.linspace(0, 10, 11)
```

```python
b = pow(a, 4) # a ** 4
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
fig = plt.figure(figsize=(10, 3))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(0, 3)
axes2.set_ylim(0, 60)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')
```

## Managing Multiple Plots Together

```python
a = np.arange(0, 10)
```

```python
b = a ** 4
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x + 6
```

```python
fig, axes  = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y) # Because axes is now 2D
axes[0][0].set_ylim(0)
axes[0][0].set_xlim(0)
axes[0][0].set_xlabel('X')
axes[0][0].set_ylabel('Y')

axes[0][1].plot(a, b)

fig.suptitle("Figure level title")
# fig.savefig('new_plots.jpeg', bbox_inches = 'tight')
fig.set_figwidth(10)

plt.tight_layout() # To ensure proper spacing between the graphs
```

### Styling

```python
x = np.arange(0, 10)

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, label='X vs X', color='purple', lw=2, ls='-', marker='o') 
# lw: line width
#ls: Linestype, the options available are: '--', '-', '-.', ':', 'steps'

axes.plot(x, x ** 2, label='X vs X^2', color='red', lw='3', ls='--', marker='+', markersize='20') 
# Color also supports hex
axes.legend(loc=(1.1, 0.5))
```
