## Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('~/Documents/Data Science Masterclass/05-Seaborn/dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 4)) # Alter seaborn using matplotlib
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='Dark2', s=60, style='level of education')
# hue is used to color the dots based on the values of some other column
# palette lets you change the color code of the graphs
# 's' lets you change the individual dot size
# style must be a catagorial feature that assigns different symbols to the data points

# to save, use the same command (plt.savefig())
```

## Distribution Plots

```python
plt.figure(figsize=(10,4))
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
plt.figure(figsize=(12, 6))
sns.set(style='white')
sns.displot(data=df, x='salary', bins=20, edgecolor='blue', kde=True)
# Use this instead of histplot()
```

```python
sns.kdeplot(data=df, x='salary')
# To just get the KDE plot
```

## Count Plot

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='level of education', hue='division')
```

## Bar Plot

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.1, 0.5))
```

## Box Plot

```python
df = pd.read_csv('~/Documents/Data Science Masterclass/05-Seaborn/StudentsPerformance.csv')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, y='math score')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, y='math score', x='parental level of education')
# Specifying 'x' helps segregate the data by some category
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, y='math score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

## Pair Plots

```python
sns.pairplot(df, hue='gender', corner=True)
# Don't call pairplot for all columns, instead, separate out the columns you need into another DF, and use pairplot on that DF instead
```

## Cat Plot

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='test preparation course')
# A more generalized way of plotting, which lets you have grids by making use of row and col parameters
```
