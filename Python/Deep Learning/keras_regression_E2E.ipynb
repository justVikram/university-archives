# Feature Engineering

```python
# Dropping useless column
df = df.drop('id', axis=1)
```

```python
# Extracting info from date column
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].apply(lambda date: date.year)
df['month'] = df['date'].apply(lambda date: date.month)

df = df.drop('date', axis=1).drop('zipcode', axis=1)

df.head()
```

```python
plt.figure(figsize=(12,6))
sbn.boxplot(data=df, x='month', y='price')
```

# Creating Model

```python
# Tensorflow needs pure numpy values

X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))

model.compile('adam', 'mse')
```

```python
model.fit(
    X_train, 
    y_train, 
    validation_data=(X_test, y_test), 
    batch_size=128,
    epochs=400
)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
plt.xlim(0,)

# Loss: Loss on training data
# Val Loss: Loss on test data

# Since both overlap, we conclude model is good and there is no overfitting.
# However, if you saw that orange line was going up, but blue was going down, it means that
# you are overfitting
```

```python
from sklearn.metrics import mean_squared_error, explained_variance_score

predictions = model.predict(X_test)

RMSE = np.sqrt(mean_squared_error(y_test, predictions))
RMSE
```

```python
df['price'].mean()
# > 20% error, which is very average
```

```python
explained_variance_score(y_test, predictions) # Value of 1 is ideal
```
